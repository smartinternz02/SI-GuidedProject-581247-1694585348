```python
import seaborn as sns
```

```python
print(sns.get_dataset_names())
```

```python
df=sns.load_dataset('car_crashes')
```

# Scatter Plot

```python
sns.scatterplot(x="speeding",y="ins_losses",data=df)
```

## Bar Plot

```python
sns.barplot(data=df,x="speeding",y="ins_losses",ci=None)
```

# Joint Plot

```python
sns.jointplot(x="alcohol",y="not_distracted",data=df)
```

# Count Plot

```python
sns.countplot(x="alcohol",data=df)
```

# Box Plot

```python
sns.boxplot(x="speeding",y="alcohol",data=df)
```

# Relplot

```python
sns.relplot(x="speeding",y="ins_losses",data=df,hue="alcohol")
```

# Line Plot

```python
sns.lineplot(x="speeding",y="alcohol",data=df,ci=None)
```

# Pie Chart

```python
import matplotlib.pyplot as plt
labels=["speeding","ins_losses","ins_premium","alcohol"]
numbers=[75,70,80,60]
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])  #[left,bottom,width,height]
axes1.pie(numbers,labels=labels,autopct="%0.2f%%",colors=["orange","pink","green","blue"])
axes1.legend()
```

# Heat Map

```python
corr=df.corr()
corr
sns.heatmap(corr,annot=True,cmap="YlGnBu")
```
